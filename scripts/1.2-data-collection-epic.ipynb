```python
from epicstore_api import EpicGamesStoreAPI, models
import pandas as pd
import json
```

```python
def fetch_all_epic_games():
    api = EpicGamesStoreAPI()
    start = 0
    batch = 100
    all_items = []
    
    while True:
        resp = api.fetch_store_games(
                count = batch,
                start=start,
                product_type = models.product_types.EGSProductType.PRODUCT_GAME,
                with_price=True,
        )

        elements = resp["data"]["Catalog"]["searchStore"]["elements"]
        if elements is None or len(elements) == 0:
            break
        else:
            all_items.extend(elements)
            start += len(elements)
            print(start)

    return all_items
```

```python
df_games = fetch_all_epic_games()
print(f"Fetched {len(df_games)} games.")
with open("epic_games_all.json", "w", encoding="utf-8") as f:
    json.dump(df_games, f, ensure_ascii=False, indent=2)
```
